# Theoretical Questions

**Q1. What is the difference between a function and a method in Python?**


**Ans:** A function is a standalone block of reusable code defined with def (or lambda) that can be called on its own. A method is a function that’s associated with an object and is called on that object, implicitly receiving the object as its first argument (self).

**Example:**

```python
# Function

def greet(name):
    return f"Hello, {name}!"
greet("Alice")
```

```python
# Method

s = "hello"
res = s.upper()   # upper() is a method of string objects
print(res)
```

**Q2. Explain the concept of function arguments and parameters in Python.**

**Ans:**

* **Parameters** are the variable names listed in a function’s definition.

* **Arguments** are the actual values we pass to the function when calling it.
We can pass arguments positionally, by keyword, unpacking with * (for sequences) or ** (for dicts), and even use default values.

**Example:**

```python
def command(text, times):         # text, times are the parameters
    print((text + "!" + " ") * times)

command("Halt", 3)                 # Halt, 3 are the arguments
```

**Q3. What are the different ways to define and call a function in Python?**


**Ans:**

* **Defining:**

1. With def (named functions)

2. With lambda (anonymous, single-expression)

3. Nested functions (functions defined inside other functions)

>    ////

* **Calling:**

1. Positional arguments

2. Keyword arguments

3. Unpacking with *args and **kwargs


**Example:**

```python
# def
def add(a, b): return a + b

# lambda
a = lambda x, y: x * y

# nested
def outer(x):
    def inner(y):
        return x + y
    return inner

print(add(2,3))
print(a(4,5))
print(outer(10)(7))
```

```python
def summarize(title, *args, **kwargs):
    print("Title:", title)
    print("Additional positional:", args)   # a tuple
    print("Additional keywords:", kwargs)    # a dict

summarize(
    "Report",
    2020, 2021,             # these go into `args`
    author="Alice",         # and this into `kwargs`
    pages=42
)
```

**Ans:** The return statement

Ends function execution

Sends back a value (or tuple of values) to the caller

**Example:**

**Q6. Explain the concept of generators in Python and how they are defined.**

**Ans:**

A generator is a special kind of iterator defined with:

* A function containing one or more yield statements, or

* A generator expression in parentheses.


When called, a generator function returns a generator object, which produces values on‐the‐fly each time we call next().

```python
def kore_made_count(n):
    i = 1
    while i <= n:
        yield i
        i += 1

for num in kore_made_count(3):
    print(num)
```

```python
sq = (x*x for x in range(7))
print(next(sq))
print(next(sq))
print(next(sq))
print(list(sq))
```

**Q7. What are the advantages of using generators over regular functions?**


**Ans:**


* **Memory efficiency:** Produce items one at a time, no need to build a full list in memory.

* **Lazy evaluation:** Computations happen only when requested.

* **Pipelining:** Easy to chain generators for stream processing.

* **Cleaner code:** Simpler stateful iteration without managing the entire sequence.

**Example:** Processing a very large file line by line without loading it entirely into memory.



```python
def gen_numbers(n):
    """Yield numbers from 1 up to n."""
    for i in range(1, n+1):
        yield i

def filter_even(nums):
    """Yield only the even numbers."""
    for x in nums:
        if x % 2 == 0:
            yield x

def square(nums):
    """Yield each number squared."""
    for x in nums:
        yield x * x

# Now we chain them: generate 1–10, keep evens, then square them
squares_of_evens = square(filter_even(gen_numbers(10)))

# Consume the generator and print each result
for sq in squares_of_evens:
    print(sq)
```

**Q10. What is the difference between map(), reduce(), and filter() functions in Python?**

* **map()** – transforms each element: one input → one output.

* **filter()** – selects elements that satisfy a predicate: input elements → subset of them.

* **reduce()** - (from functools) – aggregates all elements to a single value by cumulatively applying a binary function.

**Example:**

```python
from functools import reduce

nums = [4,7,3,16]

map_ex    = list(map(lambda x: x*2, nums))
filter_ex = list(filter(lambda x: x%2==0, nums))
reduce_ex = reduce(lambda a,b: a+b, nums)

print(map_ex)
print(filter_ex)
print(reduce_ex)
```

```python
# Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.

prime_desu = lambda n: n > 1 and all(n % i != 0 for i in range(2, int(n**0.5) + 1))
n = int(input("Enter a number (1–200) to check for prime: "))
print(f"{n} is", "prime." if prime_desu(n) else "not prime.")
```

```python
# Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
# terms.

n = int(input("Enter number of Fibonacci terms: "))

def fib(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

print("Fibonacci sequence:", list(fib(n)))
```

```python
# Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(max_exponent):

  return [2 ** i for i in range(max_exponent + 1)]

max_exponent = int(input("Enter exponent for powers of 2: "))

print("Powers of 2:", list(powers_of_two(max_exponent)))
```

```python
# Q7. Implement a generator function that reads a file line by line and yields each line as a string.


def file_line_reader(filename):
    with open(filename, 'r') as f:
        for line in f:
            yield line.rstrip('\n')

fname = input("Enter filename to read lines: ")
print("File lines:")
for line in file_line_reader(fname):
    print(line)
```
